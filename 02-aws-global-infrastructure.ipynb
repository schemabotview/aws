AWS operates one of the largest and most reliable cloud infrastructures in the world. It is built across **physical locations globally**, designed for high availability, fault tolerance, and low latency — so your applications can serve users anywhere, reliably.

![AWS Global Infrastructure](https://raw.githubusercontent.com/schemabotview/aws/main/img/aws-global-infrastructure.svg)

## Regions

A **Region** is a physical geographic area that contains AWS data centers. Each region is completely independent — it has its own power, networking, and connectivity.

AWS has **30+ regions** worldwide (and growing), spread across North America, South America, Europe, Middle East, Africa, and Asia Pacific.

Examples:
- `us-east-1` — US East (N. Virginia)
- `eu-west-1` — Europe (Ireland)
- `ap-south-1` — Asia Pacific (Mumbai)
- `ap-southeast-1` — Asia Pacific (Singapore)

### How to choose a Region

| Factor | Consideration |
|---|---|
| **Compliance & data residency** | Some regulations require data to stay within a country or region |
| **Latency** | Choose a region closest to your end users |
| **Service availability** | Not all AWS services are available in every region |
| **Pricing** | Costs vary slightly between regions |

## Availability Zones (AZs)

Each Region is divided into multiple **Availability Zones**. An AZ is one or more discrete data centers with redundant power, networking, and connectivity.

- Each region has a **minimum of 3 AZs** (most have 3–6)
- AZs within a region are physically separated by a meaningful distance (tens of miles apart) — but connected with high-bandwidth, low-latency fiber
- AZ names: `us-east-1a`, `us-east-1b`, `us-east-1c`


### Why AZs matter

By deploying your application across multiple AZs, you protect against a **single data center failure**. If one AZ goes down, your app keeps running in the others. This is the foundation of **high availability** on AWS.

## Edge Locations & Points of Presence

**Edge Locations** are AWS data centers deployed in cities around the world — separate from regions — used to deliver content closer to end users with minimal latency.

AWS has **400+ Points of Presence** (Edge Locations + Regional Edge Caches) in **90+ cities across 40+ countries**.

### What uses Edge Locations?

| Service | Purpose |
|---|---|
| **Amazon CloudFront** | CDN — caches and delivers static/dynamic content from the nearest edge |
| **Amazon Route 53** | DNS — routes users to the fastest endpoint |
| **AWS WAF** | Web Application Firewall — filters traffic at the edge |
| **AWS Shield** | DDoS protection at the edge |
| **Lambda@Edge** | Run Lambda functions at the edge, close to users |

> Edge Locations are not full regions — they don't run EC2 or RDS. They are purpose-built for **content delivery and DNS**.

## Local Zones

**Local Zones** are extensions of an AWS Region, placed in densely populated metropolitan areas that are far from the nearest region.

- They bring select AWS services (EC2, EBS, RDS, etc.) physically closer to end users
- Ideal for use cases requiring **single-digit millisecond latency** — gaming, media production, live streaming, real-time simulations
- Example: `us-east-1-bos-1` is a Local Zone in Boston, an extension of the US East (N. Virginia) region

Think of Local Zones as a mini-region attached to a parent region, without full redundancy.

## Wavelength Zones

**Wavelength Zones** embed AWS compute and storage services within telecommunications providers' 5G networks.

- Designed for applications that need **ultra-low latency over 5G** (single-digit milliseconds)
- Traffic from a 5G device never leaves the telecom network to reach AWS compute
- Use cases: connected vehicles, smart factories, AR/VR, real-time gaming
- Available with partners like Verizon, Vodafone, KDDI, SK Telecom

## AWS Outposts

**AWS Outposts** brings AWS infrastructure, services, APIs, and tools to your **on-premises data center**.

- AWS ships and installs physical rack hardware at your location
- You run AWS services locally — same APIs, same console — but in your building
- Connected back to the parent AWS region over a private network link
- Use cases: data that must stay on-premises (regulatory), ultra-low latency for local systems, hybrid workloads

Available in two form factors:
- **Outposts Rack** — full 42U rack
- **Outposts Servers** — 1U or 2U servers for space-constrained locations

## Summary

| Component | What it is | Count |
|---|---|---|
| **Region** | Independent geographic area with multiple data centers | 30+ |
| **Availability Zone** | Isolated data center cluster within a region | 3–6 per region |
| **Edge Location** | Delivery point for CloudFront, Route 53, WAF | 400+ globally |
| **Local Zone** | Region extension in a metro city for low latency | Select cities |
| **Wavelength Zone** | AWS compute embedded in 5G telecom networks | Select carriers |
| **Outposts** | AWS rack hardware deployed on-premises | Your data center |

**The key principle:** AWS separates infrastructure at multiple levels — regions for geographic isolation, AZs for fault tolerance within a region, and edge locations for performance at the network edge.
