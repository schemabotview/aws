Cloud computing is the **on-demand delivery of IT resources over the internet** with pay-as-you-go pricing.  
Instead of owning and maintaining physical data centers and servers, you can access compute, storage, databases, and dozens of other services from a cloud provider like AWS — and pay only for what you use.

> _Think of it like electricity: you don't build a power plant to power your home. You plug in and pay for what you consume._

## Traditional IT vs. Cloud

| | Traditional IT | Cloud Computing |
|---|---|---|
| **Hardware** | Buy and own servers | Rent on-demand |
| **Capacity** | Over-provision to handle peak | Scale up/down as needed |
| **Speed** | Weeks to provision | Minutes (or seconds) |
| **Cost model** | Large upfront capital (CapEx) | Pay-as-you-go (OpEx) |
| **Maintenance** | Your team manages hardware | Provider handles undifferentiated heavy lifting |
| **Global reach** | Expensive, slow | Deploy worldwide in minutes |

### The old pain points
- Paying for idle capacity (servers that sit at 10% utilisation)
- Long procurement cycles (ordering servers takes weeks)
- Guessing capacity — too little = outage, too much = wasted spend
- Maintaining physical data centers (racks, cooling, power, security)

## Cloud Deployment Models

### ☁️ Public Cloud
Resources are owned and operated by a third-party provider (AWS, Azure, GCP) and shared across customers.  
You never see or touch the hardware. This is the most common model.

### 🏢 Private Cloud
Cloud infrastructure dedicated to a single organisation — either on-premises or hosted by a provider.  
More control and isolation, but also more responsibility and cost.

### 🔀 Hybrid Cloud
A mix of public and private cloud, connected together.  
Common when organisations have legacy systems on-premises but want to leverage cloud for new workloads.

### 🤝 Multi-Cloud
Using multiple public cloud providers (e.g. AWS + GCP).  
Avoids vendor lock-in, but adds operational complexity.

## Cloud Service Models

| Model | Full Name | You manage | AWS Examples |
|---|---|---|---|
| **IaaS** | Infrastructure as a Service | OS, runtime, app, data | EC2, VPC, EBS |
| **PaaS** | Platform as a Service | App and data only | Elastic Beanstalk, RDS |
| **SaaS** | Software as a Service | Nothing (just use it) | Amazon WorkMail, Chime |

## 6 Advantages of Cloud Computing (AWS)

AWS defines these six key advantages:

1. **Trade capital expense for variable expense**  
   Pay only for what you consume instead of investing in data centers upfront.

2. **Benefit from massive economies of scale**  
   AWS aggregates usage from hundreds of thousands of customers, passing savings on to you.

3. **Stop guessing capacity**  
   Scale up or down in minutes. No more over-provisioning or under-provisioning.

4. **Increase speed and agility**  
   New resources are a click away — developers can experiment faster and iterate quicker.

5. **Stop spending money on running and maintaining data centers**  
   Focus on your customers, not the infrastructure that runs your business.

6. **Go global in minutes**  
   Deploy your application in multiple AWS Regions around the world with just a few clicks.

## AWS — The Leading Cloud Provider

Amazon Web Services (AWS) launched in **2006** and is the world's most comprehensive and broadly adopted cloud platform.  
It offers **200+ fully featured services** spanning compute, storage, databases, ML, networking, security, and more — all available from data centers globally.

Key facts:
- Available in **30+ geographic regions**, each made up of multiple Availability Zones
- Used by startups, enterprises, and governments worldwide
- Recognised as a leader in the cloud market (largest market share)

You interact with AWS via:
- **AWS Management Console** — web-based GUI
- **AWS CLI** — command-line interface
- **SDKs** — programmatic access (Python `boto3`, JavaScript, Java, etc.)
